```python
base_url = "https://classical.music.apple.com"
```

```python
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/116.0'}
response = requests.get('https://classical.music.apple.com/cn/browse/conductors', params={}, headers=headers)
response.encoding = 'UTF-8'

html = response.text
```

```python
# print(html)
```

```python
import json
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text)

# soup_output = soup.find_all('div', class_='top-search-lockup svelte-oo03qi top-search-lockup--ellipse')
soup_output = soup.find_all('div', class_='top-search-lockup__action svelte-oo03qi')
```

```python
# print(soup_output)
```

```python
conductors = []
for div in soup_output:
    a_tag = div.find('a')
    if a_tag and a_tag.has_attr('aria-label') and a_tag.has_attr('href'):
        name = a_tag['aria-label']
        url = a_tag['href']
        conductors.append({'name': name, 'url': base_url + url})

print(conductors)
```

```python
# conductors = conductors[:3]
```

```python
for conductor in conductors:
    print(f"Name: {conductor['name']}, URL: {conductor['url']}")
    response = requests.get(conductor['url'], headers=headers)
    response.encoding = 'UTF-8'
    soup = BeautifulSoup(response.text, 'html.parser')

    id = conductor['url'].split('/')[-1]
    conductor['id'] = id
    print(f"Conductor ID: {conductor['id']}")

    image = soup.find('picture', class_='svelte-10tj07c')
    image_url = image.find('source')['srcset'].split(' ')[1].split(',')[1]
    print(f"Image URL: {image_url}")
    conductor['image'] = image_url

    introduction = soup.find('div', class_='truncate-wrapper svelte-p1v1m6')
    introduction_text = introduction.get_text(strip=True) if introduction else "No introduction available"
    conductor['introduction'] = introduction_text
    print(introduction_text)

    albums = soup.find_all('div', class_='shelf-list-item svelte-1q9gr8j')
    print("Albums:")
    album_list = []
    for album in albums:
        album = album.find('span', "multiline-clamp__text svelte-1a7gcr6")
        album_name = album.get_text(strip=True)
        album_url = album.find('a')['href']
        print(f"- {album_name} - {base_url + album_url}")
        album_id = album_url.split('/')[-1]
        print(f"Album ID: {album_id}")
        album_list.append({'id': album_id, 'name': album_name, 'url': base_url + album_url, "conductor_id" : conductor['id']})
    conductor['albums'] = album_list
```

```python
albums = []
for conductor in conductors:
    for album in conductor['albums']:
        albums.append(album)

print(len(albums))
```

```python
# albums = albums[:3]
```

```python
for album in albums:
    print(f"Album Name: {album['name']}, URL: {album['url']}")
    response = requests.get(album['url'], headers=headers)
    response.encoding = 'UTF-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    
    id = album['id']
    print(f"Album ID: {id}")
    
    print(f"Conductor ID: {album['conductor_id']}")

    image = soup.find('picture', class_='svelte-10tj07c')
    image_url = image.find('source')['srcset'].split(' ')[1].split(',')[1]
    print(f"Image URL: {image_url}")
    album['image'] = image_url

    songs = soup.find_all('div', class_="section-content svelte-1cv53zo")
    song_list = []
    for song in songs:
        song_soup = song
        song = song.find_all('span', class_="svelte-fwxrdi")
        if song == []:
            continue
        # print(song)

        song_composer = song[0].get_text(strip=True)
        song_name = song[1].get_text(strip=True)
        song_url = song[1].find('a')['href']
        print(f"- {song_name} - {song_composer} - {base_url + song_url}")

        song_id = album['id'] + '-' + song_url.split('/')[-1]
        print(f"Song ID: {song_id}")

        movements = song_soup.find_all('div', class_="songs-list-row__song-name")
        movement_list = []
        for movement in movements:
            movement_name = movement.get_text(strip=True)
            print(f"  + {movement_name}")
            movement_list.append(movement_name)

        song_list.append({
            'id': song_id,
            'name': song_name,
            'album_id': album['id'],
            'conductor_id': album['conductor_id'],
            'composer': song_composer,
            'url': base_url + song_url,
            'movements': movement_list
        })
    
    album['songs'] = song_list
```

```python
songs = []
for album in albums:
    for song in album['songs']:
        songs.append(song)

print(len(songs))
print(songs)
```

```python
data = {
    "conductors": conductors,
    "albums": albums,
    "songs": songs
}



with open("classical_music_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
